```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 加载数据
sto_data = pd.read_csv('2014-2024 Buffet Portfolio - stock.csv')
sto_data.rename(columns={'TICKER': 'ticker'}, inplace=True)
```

```python
print(sto_data['RET'].head())
```

```python
print(sto_data['RET'].dtype)
```

```python

# 确保日期列为 datetime 格式
sto_data['date'] = pd.to_datetime(sto_data['date'])

# 将 RET 列转换为浮点数
sto_data['RET'] = pd.to_numeric(sto_data['RET'], errors='coerce')

# 打印 RET 列的数据类型和一些样本值
print("RET column dtype:", sto_data['RET'].dtype)
print("RET column sample values:", sto_data['RET'].head())

# 移除 RET 列中的 NaN 值
sto_data = sto_data.dropna(subset=['RET'])

# 基本统计
print("\nBasic statistics for Stock Prices (PRC), Returns (RET), and Trading Volumes (VOL):")
print(sto_data[['PRC', 'RET', 'VOL']].describe())

# 时间序列图：股价
plt.figure(figsize=(12, 12))
for ticker in sto_data['ticker'].unique():
    data = sto_data[sto_data['ticker'] == ticker]
    plt.plot(data['date'], data['PRC'], label=ticker)
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (PRC)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Stock Prices Over Time.png')
plt.show()

# 回报率分布图
plt.figure(figsize=(10, 6))
sns.histplot(sto_data['RET'], kde=True)
plt.title('Distribution of Stock Returns')
plt.xlabel('Return (RET)')
plt.savefig('Distribution of Stock Returns.png')
plt.show()

# 交易量箱线图
plt.figure(figsize=(12, 6))
sns.boxplot(x='ticker', y='VOL', data=sto_data)
plt.title('Trading Volumes by Stock')
plt.xlabel('Stock Ticker')
plt.ylabel('Volume (VOL)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Trading Volumes by Stock.png')
plt.show()

# 相关性热力图
correlation = sto_data[['PRC', 'RET', 'VOL']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Price, Return, and Volume')
plt.savefig('Correlation between Price, Return, and Volume.png')
plt.tight_layout()
plt.show()

# 数据样本
print("\nSample of the stock data:")
print(sto_data[['date', 'ticker', 'PRC', 'RET', 'VOL']].head(10))
```

```python
# 移除无穷大值
sto_data = sto_data[~np.isinf(sto_data['RET'])]

# 移除极端异常值（例如，超过 3 个标准差的值）
sto_data = sto_data[np.abs(sto_data['RET'] - sto_data['RET'].mean()) <= (3 * sto_data['RET'].std())]
```

```python
# 获取最早和最晚的日期
earliest_date = sto_data['date'].min()
latest_date = sto_data['date'].max()

# 计算时间范围
date_range = latest_date - earliest_date

# 输出结果
print(f"Data Time Range:")
print(f"Start: {earliest_date.strftime('%Y-%m-%d')}")
print(f"End: {latest_date.strftime('%Y-%m-%d')}")
print(f"Range: {date_range.days} days")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Read the data
merged_data = pd.read_csv('merged_financial_stock_macro_data_with_better_features.csv')

# Convert date column to datetime and filter the date range
merged_data['date'] = pd.to_datetime(merged_data['date'])
merged_data = merged_data[(merged_data['date'] >= '2014-07-31') & (merged_data['date'] <= '2023-12-31')]

# 1. Basic Information
print("Dataset Basic Information:")
print(merged_data.info())

# 2. Missing Value Overview
print("\nMissing Value Count and Percentage for Each Column:")
missing_summary = merged_data.isnull().sum()
missing_percent = 100 * merged_data.isnull().sum() / len(merged_data)
missing_table = pd.concat([missing_summary, missing_percent], axis=1, keys=['Missing Count', 'Missing Percentage'])
print(missing_table.sort_values('Missing Percentage', ascending=False))

# 3. Visualize Missing Values
plt.figure(figsize=(15, 8))
sns.heatmap(merged_data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap (2014-07-31 to 2023-12-31)')
plt.tight_layout()
plt.show()

# 4. Missing Value Patterns
print("\nMissing Value Patterns (Top 20):")
print(merged_data.isnull().sum().sort_values(ascending=False).head(20))

# 5. Check for Rows with No Missing Values
complete_rows = merged_data.dropna().shape[0]
print(f"\nNumber of Rows with No Missing Values: {complete_rows}")
print(f"Percentage of Complete Rows: {100 * complete_rows / len(merged_data):.2f}%")

# 6. Check Missing Value Count per Row
merged_data['missing_count'] = merged_data.isnull().sum(axis=1)
print("\nDescriptive Statistics of Missing Value Count per Row:")
print(merged_data['missing_count'].describe())

# 7. Visualize Distribution of Missing Value Count per Row
plt.figure(figsize=(10, 6))
sns.histplot(merged_data['missing_count'], kde=True, bins=30)
plt.title('Distribution of Missing Value Count per Row (2014-07-31 to 2023-12-31)')
plt.xlabel('Number of Missing Values')
plt.ylabel('Frequency')
plt.show()

# 8. Check Correlation of Missing Values
high_missing_cols = missing_table[missing_table['Missing Percentage'] > 5].index
correlation_matrix = merged_data[high_missing_cols].isnull().corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Missing Value Correlation Heatmap (Columns with >5% Missing, 2014-07-31 to 2023-12-31)')
plt.tight_layout()
plt.show()

# 9. Analyze Missing Values Over Time
missing_over_time = merged_data.set_index('date').isnull().resample('M').mean()
plt.figure(figsize=(15, 8))
for col in high_missing_cols:
    plt.plot(missing_over_time.index, missing_over_time[col], label=col)
plt.title('Missing Values Over Time (2014-07-31 to 2023-12-31)')
plt.xlabel('Date')
plt.ylabel('Proportion of Missing Values')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 10. Output Sample Rows with Missing Values
print("\nSample Rows with Missing Values (First 5 rows):")
print(merged_data[merged_data.isnull().any(axis=1)].head())

# 11. Analyze Missing Values by Stock
if 'ticker' in merged_data.columns:
    missing_by_stock = merged_data.groupby('ticker').apply(lambda x: x.isnull().mean()).mean(axis=1)
    print("\nAverage Missing Value Proportion for Each Stock:")
    print(missing_by_stock.sort_values(ascending=False).head())

    plt.figure(figsize=(12, 6))
    missing_by_stock.sort_values().plot(kind='bar')
    plt.title('Average Missing Value Proportion by Stock (2014-07-31 to 2023-12-31)')
    plt.xlabel('Stock Ticker')
    plt.ylabel('Average Missing Value Proportion')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Remove the temporary 'missing_count' column
merged_data = merged_data.drop('missing_count', axis=1)
```
